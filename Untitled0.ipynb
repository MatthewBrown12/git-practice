```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tickers = ['RR.L', 'HSBA.L', 'GSK.L', 'NWG.L', 'BARC.L', 'LLOY.L'] # .L refers to London stocks
data = yf.download(tickers, start = "2023-01-01", end = "2024-01-01")
data.head()
```

```python
data = data['Adj Close'] #Only interested in Adjusted Close data this time
```

```python
data.plot() #Plots Stock prices over the past year
```

```python
log_returns = np.log(data / data.shift(1)) #divides stock price by the day before, used log to adjust for larger prices
#.shift shifts the rows by 1
log_returns.head()
```

```python
#calculates return from portfolio over the year, as well as volatility
def portfolio_performance(weights, log_returns):
  mean_returns = log_returns.mean()
  cov_matrix = log_returns.cov()
  #weights is how much we are going to put into each stock
  portfolio_return = np.sum(mean_returns * weights) * 252 #there are 252 trading days in the year
  portfolio_volatility =  np.sqrt(np.dot(weights.T, np.dot(cov_matrix * 252, weights))) #.T is the transpose

  return portfolio_return, portfolio_volatility

def Portfolio_Sharpe(weights, log_return, riskfree = 0.1): #riskfree is the bank interest rate i.e if you don't invest
  p_return, p_volatility = portfolio_performance(weights, log_returns)
  return -(p_return - riskfree)/ p_volatility

#you want as large (in terms of absolute value) sharpe ratio as possible
```

```python
num_portfolios = 7_000 #7,000
all_weights = np.zeros(shape = (num_portfolios, len(tickers)))
returns = np.zeros(num_portfolios)
volatilities = np.zeros(num_portfolios)
sharpe_ratios = np.zeros(num_portfolios)

for i in range(num_portfolios):
  weights = np.random.random(len(log_returns.columns))
  weights /= np.sum(weights)
  all_weights[i, :] = weights
  returns[i], volatilities[i] = portfolio_performance(weights, log_returns)
  sharpe_ratios = -Portfolio_Sharpe(weights, log_returns)

```

```python
plt.figure(figsize = (10,6))
plt.scatter(volatilities, returns, c = sharpe_ratios, cmap = "viridis")
#colours each point based on value of sharpe_ratio
plt.xlabel("volatility (Risk)")
plt.ylabel("Expected Return")
plt.title("Feasible Set")
plt.show()
```
